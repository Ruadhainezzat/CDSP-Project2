# Library System

## Create a library class that is: 
     1. initialized with books
     2. can display every book available
     3. allows you to lend books
     4. allows you to return books

```python
class Library:

    def __init__(self,listofbooks):
        self.listofbooks = listofbooks

    def printMenu(self):
        listofChoices = ['Display all available books','Lend a book','Return a book','Exit']
        print(""" =============== LIBRARY MENU ================
        Enter the number of your choice:
        """)
        for c, value in enumerate(listofChoices,1):
           print(c,value)
        print("===============================================")

    def returnMenu(self):
        listofSubChoices = ['Return to Library Menu','Exit']
        print("===============================================")
        print('Enter the number of your choice:\n')
        for c, value in enumerate(listofSubChoices,1):
           print(c,value)
        print("===============================================")       

    def displayAvailablebooks(self):
        print("The books we have in our library are as follows:")
        print("===============================================")
        for c, value in enumerate(self.listofbooks,1):
           print(c,value)
        print("===============================================")
                                        
    def lendBook(self,requestedBook):
        if requestedBook.upper() in self.listofbooks:
            print('The book has been requested.') 
            self.listofbooks.remove(requestedBook.upper())
        else:
            print('The requested book is either not available or has already been taken.')
     
    def returnBook(self,returnedBook):
        self.listofbooks.append(returnedBook.upper())
        print('Thanks for returning this book.')
```

```python
# Initializing library with those books
listofbooks = ["The Last Battle","The Screwtape Letters","The Great Divorce"]
copy_listofbooks = []
counter = 0
while counter < len(listofbooks):
            copy_listofbooks.append(listofbooks[counter].upper()) 
            counter = counter + 1
library = Library(copy_listofbooks)

done = False

while not done: 
      library.printMenu()

      choice = int(input("Enter Choice:"))

      if choice == 1:
            library.displayAvailablebooks()

            library.returnMenu()
            subChoice = int(input("Enter Choice:"))
            if subChoice == 1:
                  continue
            done = True
      
      elif choice == 2:
            requestedBook = input("Book Name:")
            library.lendBook(requestedBook)

            library.returnMenu()
            subChoice = int(input("Enter Choice:"))
            if subChoice == 1:
                  continue
            done = True
      
      elif choice == 3:
            returnedBook = input("Book Name:")
            library.returnBook(returnedBook)
            
            library.returnMenu()
            subChoice = int(input("Enter Choice:"))
            if subChoice == 1:
                  continue
            done = True
      
      else:
            done = True
```
